```python
a=np.arange(-20,20,0.5)
def finda(x):
    x/=x.max()-x.min()
    x-=x.min()-1
    xp=[np.log(x) if ai==0 else x**ai for ai in a]
    a_i=np.argmin([np.abs(xp_i.skew()) for xp_i in xp])
    return a[a_i],xp[a_i].skew()
```

```python
for s in df.columns:
    print(s,finda(df[s]))
```
